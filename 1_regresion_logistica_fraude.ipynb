```python
# Modelo para clasificar transacciones como legítimas o fraudulentas
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import pandas as pd

# Preprocesamiento: elimina y transforma columnas categóricas
# (Asegúrate de ejecutar esto antes de entrenar y predecir)
df_model = df.copy()
df_model = df_model.drop(['nameOrig', 'nameDest'], axis=1)
df_model = pd.get_dummies(df_model, columns=['type'])

# Divide los datos
X = df_model.drop('isFraud', axis=1)
y = df_model['isFraud']
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entrena el modelo
model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)

# Predice sobre todo el dataset (o usa x_test para solo el test)
predicciones = model.predict(X)

# Agrega la predicción al DataFrame original
resultado = df.copy()
resultado['prediccion_fraude'] = predicciones

# 0 = legítima, 1 = fraude
print(resultado[['amount', 'type', 'isFraud', 'prediccion_fraude']].head(10))
```

```python
# Imprimir el número de transacciones con formato de miles
num_legitimas = (resultado['prediccion_fraude'] == 0).sum()
num_fraudulentas = (resultado['prediccion_fraude'] == 1).sum()
print(f"El número de transacciones legítimas son {num_legitimas:,} y de fraudulentas son {num_fraudulentas:,}")
```
